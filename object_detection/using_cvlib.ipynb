Uses Yolov3 model trained on Coco dataset

```python
#Detect objects in just 9 lines of code.
import cv2
import cvlib as cv
from cvlib.object_detection import draw_bbox
from IPython.display import Image


```

```python
def classify(img):
  path='/content/drive/MyDrive/Colab Notebooks/AILens/'
  image=cv2.imread(path+img)
  bbox, label, conf = cv.detect_common_objects(image,model='yolov3')
  print(bbox, label, conf)
  output_image = draw_bbox(image, bbox, label, conf)
  cv2.imwrite(path+"object_detection.jpg", output_image)
  display(Image(filename=path+"object_detection.jpg"))
```

```python
classify('1.jpg')
```
